## Conteúdo separado das aulas da Alura

```python
import pandas as pd
```

```python
ibge = pd.read_html('http://www.idealsoftwares.com.br/indices/ipca_ibge.html')

len(ibge)
```

## Escolhendo apenas os doze meses de 2021

```python
ibge[1][0:12]
```

## Mostrando somente as colunas desejadas

```python
s = pd.DataFrame(ibge[1][0:12], columns=['Mês/ano', 'Índice do mês(em %)', 'Índice acumuladono ano (em %)', 'Índice acumulado nos últimos 12 meses (em %)'])
```

```python
s
```

## Convertendo as colunas para float

```python
s[['Índice do mês(em %)', 'Índice acumuladono ano (em %)', 'Índice acumulado nos últimos 12 meses (em %)']] = s[['Índice do mês(em %)', 'Índice acumuladono ano (em %)', 'Índice acumulado nos últimos 12 meses (em %)']].astype(float)
```

```python
s.dtypes
```

## Dividindo a coluna, só divide uma coluna de cada vez

```python
s['Índice do mês(em %)'] = s['Índice do mês(em %)'].div(100).round(2)
```

```python
s['Índice acumuladono ano (em %)'] = s['Índice acumuladono ano (em %)'].div(10000).round(2)
```

```python
s['Índice acumulado nos últimos 12 meses (em %)'] = s['Índice acumulado nos últimos 12 meses (em %)'].div(10000).round(2)
```

```python
s
```

## Renomeando as colunas

```python
s.rename(columns={'Mês/Ano': 'Mes/ano',
                      'Índice acumuladono ano (em %)': 'Acumulado no ano (%)',
                      'Índice acumulado nos últimos 12 meses (em %)': 'Acumulado 12 meses (%)',
                      'Índice do mês(em %)': 'Mes (%)'}, inplace=True)
```

```python
s
```

## Exportando os dados do DataFrame

```python
s.to_csv('data/ipca_2021.csv', sep=';')
```
