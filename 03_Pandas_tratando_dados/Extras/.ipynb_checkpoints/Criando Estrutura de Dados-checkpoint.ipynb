```python
import pandas as pd
```

# Series

```python
data = [1, 2, 3, 4, 5]
```

```python
s = pd.Series(data)
```

```python
s
```

```python
index = ['Linha' + str(i) for i in range(5)]
```

```python
index
```

```python
s = pd.Series(data = data, index = index)
```

```python
s
```

```python
data = {'Linha' + str(i) : i + 1 for i in range(5)}
```

```python
data
```

```python
s = pd.Series(data)
s
```

```python
# Somando 2 a todos os registros
s1 = s + 2
s1
```

```python
s2 = s + s1
s2
```

```python
s3 = s2 / 2
s3
```

```python
s4 = s3 * 1.5
s4
```

# DataFrame

```python
data = [[1,2,3],
        [4,5,6],
        [7,8,9]]
data
```

```python
df1 = pd.DataFrame(data = data)
df1
```

```python
index = ['Linha' + str(i) for i in range(3)]
index
```

```python
df1 = pd.DataFrame(data=data, index=index)
df1
```

```python
cols = ['Coluna' + str(i) for i in range(3)]
cols
```

```python
df1 = pd.DataFrame(data=data, index=index, columns=cols)
df1
```

```python
df1[df1 > 0] = 'A'
df1
```

```python
data2 = [(1,2,3),
        (4,5,6),
        (7,8,9)]
data2
```

```python
df2 = pd.DataFrame(data=data2, index=index, columns=cols)
```

```python
df2[df2 > 0] = 'B'
df2
```

```python
df3 = pd.concat([df1, df2])
df3
```

```python
df3 = pd.concat([df1, df2], axis=1)
df3
```
