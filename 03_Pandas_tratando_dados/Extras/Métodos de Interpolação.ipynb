```python
import pandas as pd
```

```python
data = [0.5, None, None, 0.52, 0.54, None, None, 0.59, 0.6, None, 0.7]
s = pd.Series(data)
```

```python
s
```

```python
s.fillna(0)
```

```python
# Preenche conforme o registro anterior
# Caso o valor 1 esteja vazio e o registro 0 seja (0.50), quando rodarmos o fillna o valor 1 ser√° (0.50)
# Primeiro ao ultimo
s.fillna(method='ffill')
```

```python
# Semelhante ao 'ffill', mas faz do ultimo ao primeiro
s.fillna(method='bfill')
```

```python
# Preenchendo com uma media dos valores
s.fillna(s.mean())
```
