## Seleções e Frequências

```python
import pandas as pd
```

```python
dados = pd.read_csv('data/aluguel_residencial.csv', sep=';')
```

```python
dados.head(10)
```

```python
# Selecione somente os imóveis classificados com tipo 'Apartamento'
selecao = dados['Tipo'] == 'Apartamento'
ap1 = dados[selecao].shape[0]
ap1
```

```python
# Selecione os imóveis classificados com tipo 'Casa', 'Casa de Condomínio' e 'Casa de Vila'
selecao2 = (dados['Tipo'] == 'Casa') | (dados['Tipo'] == 'Casa de Condomínio') | (dados['Tipo'] == 'Casa de Vila')
ap2 = dados[selecao2].shape[0]
ap2
```

```python
# Selecione os imóveis com área entre 60 e 100 metros quadrados, incluindo os limites
selecao3 = (dados['Area'] >= 60) & (dados['Area'] <= 100)
ap3 = dados[selecao3].shape[0]
ap3
```

```python
# Selecione os imóveis que tenham pelo menos 4 quartos e aluguel menor que R$ 2.000,00
selecao4 = (dados['Quartos'] >= 4) & (dados['Valor'] < 2000)
ap4 = dados[selecao4].shape[0]
ap4
```
